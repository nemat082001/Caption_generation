```python
import pandas as pd
import re
# Read the Excel file
input_file = '3DPrinted Gun with Bulet.xlsx'  
df = pd.read_excel(input_file)
df.shape
```

```python
filenames = df['Xray image']
def extract_numbers(part):
    match = re.search(r'\d+[a-zA-Z]*', part)
    return match.group() if match else ''
processed_data = []
for filename in filenames:
    words_list = filename.split('_')
    words_list[-1] = words_list[-1].split('.')[0]  
    numbers_list = [extract_numbers(part) for part in words_list]
    processed_data.append(numbers_list)

for words in processed_data:
    print(words)
```

```python
A_data=[ x[1:] for x in processed_data]
```

```python
A_data=[['1']+ x for x in A_data]
```

```python
A_data
```

```python
import random
picture_to_caption=[]
list_of_verbs=["showing","displaying","illustrating","showcasing"]
list_of_location=["placed", "positioned"]
for items in A_data:
    
        var_name="threat_"+str(items[0])
        threat_lst=globals()[var_name]
        
        th=random.choice(threat_lst)
        threat =th
        
        var_name0="bag_"+str(items[1])
        var_name1= "LEVEL_" + str(items[2]) + "_conc_" + str(items[3])
        var_name2 = "Location_" + str(items[4])
        var_name3 = "Phi_" + str(items[5])
        var_name4 = "Theta_" + str(items[6])
        var_name5= "Picture_"+str(items[7])
        caption = f"An X-ray image of a " +globals()[var_name0] 
        caption += f" {random.choice(list_of_verbs)} threat items, specifically "

        caption+=th
        
        caption += f' {random.choice(list_of_location)} '+ globals()[var_name3]   
            # caption +=' and phi '+ globals()[var_name4]
        caption +=', at the ' + globals()[var_name2] +' of the bag along with ' 
        var_name11=globals()[var_name1]
        caption +=var_name11
        caption+='.'
        print(caption)
        picture_to_caption.append(caption)
        
```
